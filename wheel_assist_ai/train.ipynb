```python
from ultralytics import YOLO
import torch
import os
import time
import pandas as pd
from datetime import datetime
```

```python
DATA_YAML = "YOLO_Obstacle_Final_lowres/data.yaml"

# BEST BALANCE FOR PHONE + TFLITE
MODEL_NAME = "yolov8n.pt"

PROJECT_NAME = "lightweight_wheelchair_obstacle_detection"

RUN_NAME = "yolov8n_obstacle"

EPOCHS = 100

IMAGE_SIZE = 240

BATCH_SIZE = 16

CONF_THRESHOLD = 0.4
```

```python
if torch.cuda.is_available():

    device = 0
    device_name = torch.cuda.get_device_name(0)
    print(device_name)

elif torch.backends.mps.is_available():

    device = "mps"
    device_name = "Apple Silicon GPU"

else:

    device = "cpu"
    device_name = "CPU"
```

```python
print("=" * 60)
print("DEVICE INFO")
print("=" * 60)

print("Using device:", device)
print("Device name:", device_name)

print("=" * 60)
```

```python
model = YOLO(MODEL_NAME)
train_start = time.time()
```

```python
results = model.train(

    # Dataset
    data=DATA_YAML,

    # Training
    epochs=EPOCHS,
    imgsz=IMAGE_SIZE,
    batch=BATCH_SIZE,

    # Device
    device=device,

    # Optimizer
    optimizer="AdamW",
    lr0=0.001,

    # Workers
    workers=0,

    # Validation
    val=True,

    # Save
    save=True,
    save_period=5,

    # Verbose
    verbose=True,

    # TensorBoard
    plots=True,

    # Project
    project=PROJECT_NAME,
    name=RUN_NAME,

    # Augmentations
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=10,
    translate=0.1,
    scale=0.3,
    shear=2,

    fliplr=0.5,
    mosaic=1.0,
    mixup=0.1,

    # Early stopping
    patience=50
)
```

```python
train_end = time.time()

train_time = (train_end - train_start) / 60
```

```python
BEST_MODEL = os.path.join(
    PROJECT_NAME,
    RUN_NAME,
    "weights",
    "best.pt"
)

print("\nBest model path:")
print(BEST_MODEL)
```

```python
best_model = YOLO(BEST_MODEL)
print("\nRunning validation...")

val_metrics = best_model.val()
```

```python
print("\nRunning test evaluation...")

test_metrics = best_model.val(
    data=DATA_YAML,
    split="test"
)
```

```python
print("\n" + "=" * 60)
print("FINAL TEST METRICS")
print("=" * 60)

print(f"mAP50: {test_metrics.box.map50:.4f}")
print(f"mAP50-95: {test_metrics.box.map:.4f}")

print(f"Precision: {test_metrics.box.mp:.4f}")
print(f"Recall: {test_metrics.box.mr:.4f}")
```

```python
print("\nRunning inference benchmark...")

sample_image = "YOLO_Obstacle_Final/test/images"

start_time = time.time()

best_model.predict(
    source=sample_image,
    conf=CONF_THRESHOLD,
    verbose=False
)

end_time = time.time()

inference_time = end_time - start_time

fps = 1 / inference_time if inference_time > 0 else 0

print(f"Inference Time: {inference_time:.4f} sec")
print(f"Approx FPS: {fps:.2f}")
```

```python
metrics_txt = os.path.join(
    PROJECT_NAME,
    RUN_NAME,
    "final_metrics.txt"
)

with open(metrics_txt, "w") as f:

    f.write("=" * 60 + "\n")
    f.write("WHEELCHAIR OBSTACLE DETECTION RESULTS\n")
    f.write("=" * 60 + "\n\n")

    f.write(f"Date: {datetime.now()}\n\n")

    f.write(f"Model: {MODEL_NAME}\n")
    f.write(f"Epochs: {EPOCHS}\n")
    f.write(f"Image Size: {IMAGE_SIZE}\n")
    f.write(f"Batch Size: {BATCH_SIZE}\n")
    f.write(f"Device: {device_name}\n")

    f.write(f"\nTraining Time (minutes): {train_time:.2f}\n")

    f.write(f"\nInference Time (sec): {inference_time:.4f}\n")
    f.write(f"Approx FPS: {fps:.2f}\n")

    f.write("\n" + "=" * 60 + "\n")
    f.write("VALIDATION METRICS\n")
    f.write("=" * 60 + "\n")

    f.write(f"mAP50: {val_metrics.box.map50:.4f}\n")
    f.write(f"mAP50-95: {val_metrics.box.map:.4f}\n")
    f.write(f"Precision: {val_metrics.box.mp:.4f}\n")
    f.write(f"Recall: {val_metrics.box.mr:.4f}\n")

    f.write("\n" + "=" * 60 + "\n")
    f.write("TEST METRICS\n")
    f.write("=" * 60 + "\n")

    f.write(f"mAP50: {test_metrics.box.map50:.4f}\n")
    f.write(f"mAP50-95: {test_metrics.box.map:.4f}\n")
    f.write(f"Precision: {test_metrics.box.mp:.4f}\n")
    f.write(f"Recall: {test_metrics.box.mr:.4f}\n")

print("\nMetrics saved at:")
print(metrics_txt)
```

```python
csv_path = os.path.join(
    PROJECT_NAME,
    RUN_NAME,
    "results.csv"
)

df = pd.read_csv(csv_path)

summary_path = os.path.join(
    PROJECT_NAME,
    RUN_NAME,
    "training_summary.csv"
)

df.to_csv(summary_path, index=False)

print("\nTraining CSV summary saved.")
```
